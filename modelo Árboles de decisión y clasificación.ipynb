```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Generamos datos de ejemplo
np.random.seed(0)
X = np.random.randn(100, 2)  # Variables independientes (características)
y = (X[:, 0] + X[:, 1] > 0).astype(int)  # Variable dependiente (objetivo)

# Visualizamos los datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.title("Gráfico de Dispersión de los Datos de Ejemplo")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Creamos un modelo de Árbol de Decisión para clasificación
modelo = DecisionTreeClassifier()

# Entrenamos el modelo con los datos de entrenamiento
modelo.fit(X_train, y_train)
```

```python
# Hacemos predicciones usando el modelo entrenado
y_pred = modelo.predict(X_test)
```

```python
# Calculamos la precisión del modelo en el conjunto de prueba
precision = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {precision}")

# Creamos una matriz de confusión para evaluar el rendimiento
matriz_confusion = confusion_matrix(y_test, y_pred)
print("Matriz de Confusión:")
print(matriz_confusion)
```
